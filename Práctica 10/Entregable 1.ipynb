```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Obtención de los datos

```python
"""
celsius = np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit = np.array([-40,14,32,46,59,72,100], dtype=float)
"""
df = pd.read_csv("celsius_a_fahrenheit.csv")
df.head()
```

```python
celsius = df["Celsius"].values
fahrenheit = df["Fahrenheit"].values
```

```python
plt.scatter(celsius, fahrenheit)
plt.title("Correlación entre Celcius y Fahrenheit")
plt.xlabel("Celcius")
plt.ylabel("Fahrenheit")
plt.show()
```

```python
sns.scatterplot(x='Celsius', y="Fahrenheit", data=df)
plt.title('Correlación Celsius vs Fahrenheit')
plt.xlabel('Celsius')
plt.ylabel('Fahrenheit')
plt.show()
```

### Selección de las varibales independiente y objetivo

```python
X_train = celsius
Y_train = fahrenheit
```

### Creación de modelo

```python
print("Creando el modelo")
model = tf.keras.Sequential()
model.add(tf.keras.layers.Input(shape=(1,)))
model.add(tf.keras.layers.Dense(units=1))
```

### Compilación del modelo

```python
print("Compilando el modelo")
model.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss="mean_squared_error")
```

### Entrenamiento del modelo

```python
print("Entrenando el modelo")
epochs_hist = model.fit(X_train,Y_train,epochs=100)
```

```python
print("Evaluando el modelo entrenado")
print("Keys: ")
print(epochs_hist.history.keys())

plt.plot(epochs_hist.history["loss"])
plt.title("Progreso de Pérdida durante Entrenamiento del Modelo")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend(["Training Loss"])
plt.show()
```

```python
Temp_C = 100
Temp_C_array = np.array([[Temp_C]])
Temp_F = model.predict(Temp_C_array)
print("Temperatura de Prediccion: " + str(Temp_F))
```

### Modificacion (4)

```python
print("Creando el modelo")
model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.Input(shape=(1,)))
model2.add(tf.keras.layers.Dense(units=1))
print("Compilando el modelo")
model2.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss="mean_squared_error")
epochs_hist2 = model2.fit(X_train,Y_train,epochs=100)
print("Evaluando el modelo entrenado")
print("Keys: ")
print(epochs_hist2.history.keys())

plt.plot(epochs_hist2.history["loss"])
plt.title("Progreso de Pérdida durante Entrenamiento del Modelo")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend(["Training Loss"])
plt.show()
```

### Modificación 5 y 6

```python
print("Creando el modelo")

model3 = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=3),
    tf.keras.layers.Dense(units=5),
    tf.keras.layers.Dense(units=1)
])

print("Compilando el modelo")
model3.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss="mean_squared_error")

X_train_r = X_train.reshape(-1, 1)
Y_train_r = Y_train.reshape(-1, 1)

epochs_hist3 = model3.fit(X_train_r, Y_train_r, epochs=100)

print("Evaluando el modelo entrenado")
print("Keys: ", epochs_hist3.history.keys())

plt.plot(epochs_hist3.history["loss"])
plt.title("Progreso de Pérdida durante Entrenamiento del Modelo")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend(["Training Loss"])
plt.show()
```
