```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n_sample = 100000
ne = np.random.choice(["Bajo", "Medio", "Alto"], size=n_sample, p=[0.1, 0.3, 0.6])
eq = np.random.choice(["Basico", "Intermedio", "Avanzado"], size=n_sample, p=[0.1, 0.2, 0.7])
h = np.random.choice(["Bajas", "Medias", "Altas"], size=n_sample, p=[0.25, 0.35, 0.4])
exito = np.zeros(n_sample)
```

```python
for i in range(n_sample):
    # High performance combination
    if ne[i] == 'Alto' and eq[i] == 'Avanzado' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.85, 0.15])
    
    # Good combinations
    elif ne[i] == 'Alto' and eq[i] == 'Avanzado' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.75, 0.25])
    elif ne[i] == 'Alto' and eq[i] == 'Intermedio' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.70, 0.30])
    elif ne[i] == 'Medio' and eq[i] == 'Avanzado' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.68, 0.32])
    
    # Medium combinations
    elif ne[i] == 'Alto' and eq[i] == 'Avanzado' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.60, 0.40])
    elif ne[i] == 'Alto' and eq[i] == 'Intermedio' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.55, 0.45])
    elif ne[i] == 'Medio' and eq[i] == 'Avanzado' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.52, 0.48])
    elif ne[i] == 'Alto' and eq[i] == 'Basico' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.50, 0.50])
    elif ne[i] == 'Medio' and eq[i] == 'Intermedio' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.48, 0.52])
    
    # Lower-medium combinations
    elif ne[i] == 'Alto' and eq[i] == 'Intermedio' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.45, 0.55])
    elif ne[i] == 'Medio' and eq[i] == 'Avanzado' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.42, 0.58])
    elif ne[i] == 'Alto' and eq[i] == 'Basico' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.38, 0.62])
    elif ne[i] == 'Medio' and eq[i] == 'Intermedio' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.35, 0.65])
    
    # Low combinations
    elif ne[i] == 'Bajo' and eq[i] == 'Avanzado' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.40, 0.60])
    elif ne[i] == 'Medio' and eq[i] == 'Basico' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.32, 0.68])
    elif ne[i] == 'Alto' and eq[i] == 'Basico' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.28, 0.72])
    elif ne[i] == 'Medio' and eq[i] == 'Intermedio' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.25, 0.75])
    elif ne[i] == 'Bajo' and eq[i] == 'Avanzado' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.30, 0.70])
    elif ne[i] == 'Medio' and eq[i] == 'Basico' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.22, 0.78])
    
    # Very low combinations
    elif ne[i] == 'Bajo' and eq[i] == 'Intermedio' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.28, 0.72])
    elif ne[i] == 'Bajo' and eq[i] == 'Avanzado' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.20, 0.80])
    elif ne[i] == 'Medio' and eq[i] == 'Basico' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.15, 0.85])
    elif ne[i] == 'Bajo' and eq[i] == 'Intermedio' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.18, 0.82])
    elif ne[i] == 'Bajo' and eq[i] == 'Basico' and h[i] == 'Altas':
        exito[i] = np.random.choice([1, 0], p=[0.15, 0.85])
    
    # Worst combinations
    elif ne[i] == 'Bajo' and eq[i] == 'Intermedio' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.10, 0.90])
    elif ne[i] == 'Bajo' and eq[i] == 'Basico' and h[i] == 'Medias':
        exito[i] = np.random.choice([1, 0], p=[0.08, 0.92])
    elif ne[i] == 'Bajo' and eq[i] == 'Basico' and h[i] == 'Bajas':
        exito[i] = np.random.choice([1, 0], p=[0.05, 0.95])
    
    # Default case (shouldn't happen, but just in case)
    else:
        exito[i] = np.random.choice([1, 0], p=[0.30, 0.70])
```

```python
# Create DataFrame with all variables
data = pd.DataFrame({
    'Nivel de Experiencia': ne,
    'Equipamento': eq,
    'Habilidades': h,
    'Exito': exito.astype(int)
})
```

```python
prob_exito = data.groupby(["Nivel de Experiencia", "Equipamento", "Habilidades"])["Exito"].mean().reset_index()
```

```python
# Combine las tres variables en una sola etiqueta para el eje x
prob_exito['Etiqueta'] = prob_exito['Nivel de Experiencia'] + ' - ' + prob_exito['Equipamento'] + ' - ' + prob_exito['Habilidades']

# Graficar las probabilidades
fig, ax = plt.subplots(figsize=(16, 6))  # Figura ms ancha para 27 combinaciones
prob_exito.plot(kind='bar', x='Etiqueta', y='Exito', ax=ax, legend=False, color='steelblue')
ax.set_ylabel('P(xito = 1)')
ax.set_xlabel('Combinacin (NE - Equipamiento - Habilidades)')
ax.set_title('Probabilidades Condicionales de xito')
plt.xticks(rotation=90, ha='right')  # Rotacin vertical para mejor legibilidad
plt.tight_layout()
plt.savefig('probabilidades_condicionales_exito.png', dpi=300, bbox_inches='tight')
plt.show()
```
