```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
diabetes_dataset = load_diabetes()
```

```python
x = diabetes_dataset.data[:,2].reshape(-1,1)
y = diabetes_dataset.target
X_train = x[:-20]
X_test = x[-20:] 
Y_train = y[:-20] 
Y_test = y[-20:]
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Coeficientes: \n', regressor.coef_)
print('Independent term: \n', regressor.intercept_)
print('Mean squared error: %.2f' % mean_squared_error(Y_test, y_pred))
print("r2 score: %.2f" % r2_score(Y_test,y_pred))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].scatter(X_train, Y_train)
ax[0].plot(X_train, regressor.predict(X_train), c="g")
ax[0].set_title("Regresin lineal (train)")
ax[0].set_ylabel("target")
ax[1].scatter(X_test, Y_test)
ax[1].plot(X_test, y_pred, c="g")
ax[1].set_title("Regresin lineal (test)")
ax[1].set_xlabel("BMI")
ax[1].set_ylabel("target")
plt.suptitle("Modelo de regresin lineal")
plt.show()
```

Podemos concluir que existe una regresin entre BMI y la progresin de la diabetes, sin embargo debido al erro alto y al r^2 podemos llegar a la conclusin de que aunque relevante, no es completamente precisio el predecir la progresin de la diabetes nicamente con el BMI
