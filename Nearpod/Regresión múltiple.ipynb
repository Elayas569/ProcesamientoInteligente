# Regresión Lineal Múltiple

1. Determinal las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de intesección

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importar datos
df = pd.read_csv("Regresión múltiple.csv")
df.head()
```

### 1. Determinar las variables independientes
x1 = calls, x2 = driven

```python
X = df[['Calls','Driven']]
```

### 2. Determinar las variables dependientes Y
y = commissions

```python
Y = df[['Commissions']]
```

```python
#Ajustar el model de regresión multiple
model = LinearRegression()
model.fit(X, Y) #Entrena el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(Y, y_pred)
r2 = r2_score(Y, y_pred)
```

### 3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ', model.coef_)
print("Error cuadrático medio ", mse)
print("Varianza: ", r2)
```

Con base en el análisis de regresión múltiple, se concluye que las comisiones de los agentes dependen casi exclusivamente del número de llamadas realizadas (coeficiente de 0.63), mientras que las millas recorridas no representan un factor significativo (coeficiente de 0.01). No existe una interacción relevante entre ambas variables, por lo que la estrategia más eficiente para incrementar comisiones es enfocarse en el volumen de llamadas antes que en los desplazamientos físicos.
