```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline

#Librerias para operaciones de regresion lineal
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("./CopierSalesofAmerica.csv")
```

```python
#Determina la cantidad de filas y de columnas
data.shape
```

```python
#muestra la tabla
data.head()
```

```python
#selecciona los datos para determinar el coeficiente de correlación
X_train = np.array(data['llamadas']).reshape(-1,1)
Y_train = np.array(data['ventas'])
```

```python
#Calcular la formula de regresion lineal , coeficiente de correlación lineal
r = linear_model.LinearRegression()
#Entrenamos al modelo con los datos del entrenamiento
r.fit(X_train, Y_train)

#obtener predicciones
y_pred = r.predict(X_train)

print('Coeficiente de correlación: ' , r.coef_)
print('Interseccion: ' , r.intercept_)
print('Error cuadrático de la media: ', mean_squared_error(Y_train,y_pred))
print('Varianza: ' , r2_score(Y_train, y_pred))
```

```python
x = X_train.ravel()
plt.figure(figsize=(8,5))
plt.scatter(x,Y_train,color='steelblue', s=80, label = 'Datos')
plt.plot(x,y_pred, color = 'red',linewidth=3,label='Pendiente')
plt.xlabel('No. de llamadas')
plt.ylabel('Copiadoras vendidas')
plt.title('Practica 14 enero')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
