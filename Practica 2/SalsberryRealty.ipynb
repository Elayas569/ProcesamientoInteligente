# Regresión Lineal Múltiple - SalsBerry Realty

1. Determinal las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de intesección
4. Estimar el corto de calefacción para una Temperatura externa media de 30°, 5 pulfadas de aislamiento y un calentador con antigüedad de 10 años.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#importar datos
df = pd.read_csv("SalsberryRealty.csv")
df.head()
```

### 1. Determinar las variables independientes
x1 = temp, x2 = insul, x3 = age

```python
X = df[['Temp','Insul','Age']]
```

### 2. Determinar las variables dependientes Y
y = cost

```python
Y = df[['Cost']]
```

```python
#Ajustar el model de regresión multiple
model = LinearRegression()
model.fit(X, Y) #Entrena el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas
mse = mean_squared_error(Y, y_pred)
r2 = r2_score(Y, y_pred)
```

### 3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ', model.coef_)
print("Error cuadrático medio ", mse)
print("Varianza: ", r2)
```

### 4. Estimar para 30°, aislamiento 5", antigüedad 10 años

```python
y = model.predict([[30,5,10]])
print("El costo de calefación estimado es: ", y)
```
